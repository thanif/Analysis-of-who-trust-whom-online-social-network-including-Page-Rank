# Problem 2

```python
def Page_Rank(G, beta, eta):
    M = nx.stochastic_graph(G,copy=True)
    r_old = dict.fromkeys(M, 1.0 / G.number_of_nodes())
    i = 0
    while i<100:
        r_new = dict.fromkeys(r_old.keys(), 0) 
        for n in r_new:  
            if M.out_degree(n, weight='weight') != 0.0:
                for nbr in M[n]: 
                    r_new[nbr] += beta * r_old[n] * M[n][nbr]['weight']
            else:
                for nbr in M[n]: 
                    r_new[nbr] += 0
            S = np.sum(r_new.values())        
            r_new[n] +=  (1-S)/float(G.number_of_nodes())
        err = sum([abs(r_new[n] - r_old[n]) for n in r_new]) 
        print " i : ",i," error : ", err
        if err < eta: 
            return r_new
        i+=1
        r_old = r_new    
    print "Power Iteration failed to converge"     
  
```

```python
G = nx.read_edgelist("soc-Epinions1.txt",create_using=nx.DiGraph(), nodetype = int)
```

```python
beta = input("Please enter the teleportation probability : ")
```

```python
eta = input("Please enter the termination criteria : ")
```

```python
r = Page_Rank(G, beta, eta)
```

```python
top_10 = []
for i in range(10):
    m = max(temp.iteritems(), key=operator.itemgetter(1))[0]
    top_10.append([m,temp[m]])
    temp.pop(m)
```
